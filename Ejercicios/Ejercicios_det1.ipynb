### Ejercicio 1

Una presa perfectamente cúbica con una profundidad de 5.5 $m$ y una base de 100 $m^2$ recibe una entrada constante de líquido de 5  $m^3/hora$ y deja salir constantemente 3  $m^3/hora$. Dado que el nivel de agua inicial de la presa es de 4 $m$: 

**(a)** Indicar si la presa se desbordará y en qué momento. 

**(b)** Tras un análisis más detallado, se comprueba que la presa pierde líquido en función del nivel que alcance, estimándose en 1 $m^3$ por hora y metro de nivel. Indicar en este caso si la presa se desbordará y cuál será el nivel máximo alcanzado por la presa. 

**(c)** Resolver analíticamente el modelo y comprobar que la repuesta a las preguntas anteriores dadas por la simulación coincide con la analítica. 


La presa inevitablemente se desbordará ya que la entrada de agua es mayor que la salida, y dado que es constante la entrada de agua a la presa pues esta se desbordará en algún momento.

Superficie (S)= 100 m^2
Entrada de agua (fin) = 5 m^3/hora
Salida de agua (fout) = 3 m^3/hora
Altura presa (hmax) = 5,5 m

Altura agua inicial (h0) = 4 m

Variable de estado:
    ht= altura del líquido en tiempo t
    x = volumen del líquido en tiempo t

x0 = h0 * S'

x_t+1 = x_t + fin - font 



```python
import matplotlib.pyplot as plt

def simular_presa_A(S, fin, font, hmax, h0, Delta=1/60):
    """
    Simula la evolución del nivel de agua en una presa.
    
    Parámetros:
    - S: Área de la superficie en m^2
    - fin: Caudal de entrada de agua en m^3/hora
    - font: Caudal de salida de agua en m^3/hora
    - hmax: Altura máxima de la presa en metros
    - h0: Altura inicial del agua en metros
    - Delta: Incremento de tiempo en horas (por defecto es 1/60, equivalente a 1 minuto)
    
    Retorna:
    - tiempos: Lista de tiempos en horas
    - alturas: Lista de alturas de agua correspondientes
    - overflow_time: Tiempo en el que la presa se desborda (si ocurre)
    - overflow_height: Altura en el momento del desbordamiento (si ocurre)
    """
    ht = h0  # Altura inicial del agua
    t = 0  # Tiempo inicial
    tiempos = []
    alturas = []

    while ht < hmax:
        t += 1
        x = ht * S
        x += (fin - font) 
        ht = x / S
        tiempos.append(t)
        alturas.append(ht)

    overflow_time = None
    overflow_height = None
    
    # Verificación de desbordamiento
    if ht >= hmax:
        overflow_time = t
        overflow_height = ht
        print(f"La presa se desbordará en el tiempo {overflow_time} horas con una altura de {round(overflow_height, 4)} metros.")
    else:
        print("La presa no se desbordará.")

    # Gráfica
    plt.figure(figsize=(10, 6))
    plt.plot(tiempos, alturas)
    plt.axhline(y=hmax, color='r', linestyle='--', label='Altura máxima (5.5 m)')
    plt.xlabel('Tiempo (horas)')
    plt.ylabel('Altura de agua (m)')
    plt.title('Evolución del nivel de agua en la presa')
    plt.grid(True)
    plt.legend()
    plt.show()

    return tiempos, alturas, overflow_time, overflow_height

# Ejemplo de uso
S = 100  # Área de superficie en m^2
fin = 5  # Caudal de entrada de agua en m^3/hora
font = 3  # Caudal de salida de agua en m^3/hora
hmax = 5.5  # Altura máxima de la presa en metros
h0 = 4  # Altura inicial del agua en metros

tiempos, alturas, overflow_time, overflow_height = simular_presa_A(S, fin, font, hmax, h0)

```

### Probar a graficar con diferentes alturas como se queda la presa

```python
import matplotlib.pyplot as plt

# Definimos la función
def momento_y_nivel_maximo(S, fin, hmax, h0):
    t = 0
    x0 = h0 * S
    nivel_maximo = h0
    tiempos = []
    niveles = []

    while t < 1000:  # Limite temporal para evitar bucles infinitos
        t += 1
        fout = 1 * (x0 / S)  # 1 m^3/h por metro de nivel

        x0 = x0 + fin - fout
        h_actual = x0 / S

        tiempos.append(t)
        niveles.append(h_actual)

        if h_actual > nivel_maximo:
            nivel_maximo = h_actual

        if h_actual >= hmax:
            return tiempos, niveles, f"La presa se desbordará a las {t} horas, con una altura de {hmax} metros."

        if fout >= fin:
            break

    return tiempos, niveles, f"La presa no se desbordará. El nivel máximo alcanzado será de {nivel_maximo} metros."


# Lista de alturas iniciales
alturas_iniciales = [3, 4, 4.5, 5, 6, 7]
colores = ['b', 'g', 'r', 'c', 'm', 'y']

plt.figure(figsize=(10, 6))

# Generamos la gráfica para cada altura inicial
for i, h0 in enumerate(alturas_iniciales):
    tiempos, niveles, mensaje = momento_y_nivel_maximo(100, 5, 5.5, h0)
    plt.plot(tiempos, niveles, label=f"Altura inicial {h0} m", color=colores[i], linestyle="-")

# Añadimos la línea del nivel máximo permitido
plt.axhline(y=5.5, color='k', linestyle='--', label="Nivel máximo permitido (5.5 m)")

# Configuración de la gráfica
plt.title("Evolución del nivel de agua en la presa para distintas alturas iniciales")
plt.xlabel("Tiempo (horas)")
plt.ylabel("Nivel de agua (m)")
plt.grid(True)
plt.show()
```

## Ejercicio 2

Una población de bacterias sigue un crecimiento dado por un [modelo logístico](https://mathinsight.org/bacteria_growth_logistic_model). Entonces: 

**(a)**  Modelar la evolución de la población de las bacterias siguiendo un patrón de crecimiento constante, exponencial y logístico. 

**(b)** Calibrar los parámetros del modelo en los tres casos anteriores.

**(c)** Validar el modelo, eligiendo el que mejor se ajuste a los datos.  

**(d)** Estudiar la capacidad de carga o el máximo número de bacterias a largo plazo. 

**(e)** Estudiar el modelo asumiendo diferentes unidades de incremento del tiempo. 

**(f)** Resolver analíticamente el modelo. 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos de la tabla
tiempo = np.array([0, 16, 32, 48, 64, 80, 96, 112, 128, 144, 160])
densidad = np.array([0.022, 0.036, 0.060, 0.101, 0.169, 0.266, 0.360, 0.510, 0.704, 0.827, 0.928])

# Modelo de crecimiento exponencial
def modelo_exponencial(t, a, b):
    
    return a * np.exp(b * t)

# Modelo de crecimiento logístico
def modelo_logistico(t, K, r, P0):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Modelo de crecimiento constante simple (P0 + r * t)
def modelo_constante(t, r):
    P0=densidad[0]
    return P0 + r * t

# Ajustar el modelo exponencial a los datos
popt_exp, _ = curve_fit(modelo_exponencial, tiempo, densidad, p0=[0.02, 0.01])

# Ajustar el modelo logístico a los datos
popt_log, _ = curve_fit(modelo_logistico, tiempo, densidad, p0=[1, 0.02, 0.01])

# Ajustar el modelo de crecimiento constante simple
popt_const, _ = curve_fit(modelo_constante, tiempo, densidad, p0=[0.01])

# Calcular los valores ajustados
tiempo_fino = np.linspace(0, 160, 100)
densidad_exp_ajustada = modelo_exponencial(tiempo_fino, *popt_exp)
densidad_log_ajustada = modelo_logistico(tiempo_fino, *popt_log)
densidad_const_ajustada = modelo_constante(tiempo_fino, *popt_const)

# Representación gráfica
plt.figure(figsize=(10, 6))
plt.scatter(tiempo, densidad, label="Datos experimentales", color="black")
plt.plot(tiempo_fino, densidad_exp_ajustada, label="Modelo exponencial", color="blue")
plt.plot(tiempo_fino, densidad_log_ajustada, label="Modelo logístico", color="green")
plt.plot(tiempo_fino, densidad_const_ajustada, label="Crecimiento constante (recta)", color="red")
plt.xlabel("Tiempo (min)")
plt.ylabel("Densidad de población")
plt.title("Ajuste de modelos de crecimiento bacteriano")
plt.grid(True)
plt.legend()
plt.show()

# Resultados de los parámetros ajustados
print("Parámetros ajustados exponencial (a, b): ", popt_exp)
print("Parámetros ajustados logístico (K, r, P0): ", popt_log)
print("Parámetros ajustados constante (r): ", popt_const)

# Calcular valores ajustados para los tiempos originales
densidad_exp_pred = modelo_exponencial(tiempo, *popt_exp)
densidad_log_pred = modelo_logistico(tiempo, *popt_log)
densidad_const_pred = modelo_constante(tiempo, *popt_const)

# Función para calcular el MAE (Mean Absolute Error)
def calcular_mae(y_obs, y_pred):
    mae = np.mean(np.abs(y_obs - y_pred))
    return mae

# Calcular MAE para cada modelo
mae_exp = calcular_mae(densidad, densidad_exp_pred)
mae_log = calcular_mae(densidad, densidad_log_pred)
mae_const = calcular_mae(densidad, densidad_const_pred)

print("MAE para modelo exponencial: ", mae_exp)
print("MAE para modelo logístico: ", mae_log)
print("MAE para modelo constante: ", mae_const)
```
